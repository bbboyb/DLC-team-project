```r
# Modelo de regresión (grado2 y grado3)
coefficients_matrix_pol2 <- function(x0,y){
    x <- cbind(1,x0,x0^2)
    return(solve(t(x)%*%x)%*%t(x)%*%y)
}
coefficients_matrix_pol3 <- function(x0,y){
    x <- cbind(1,x0,x0^2,x0^3)
    return(solve(t(x)%*%x)%*%t(x)%*%y)
}
```

```r
# Regresión: Francia
y <- data[1:9,3] # Y es la variable objetivo: Emisión de NOx
x <- data[1:9,5] # X es la variable independiente: Distancia recorrida

# Parámetros para estandarizar los datos
x.mean <- mean(x)
x.sd <- sd(x)
y.mean <- mean(y)
y.sd <- sd(y)
x.scaled <- scale(x)
y.scaled <- scale(y)

# Coeficientes de la regresión:
coef.pol3 <- coefficients_matrix_pol3(x.scaled,y.scaled)
plot(x,y,pch=16,ylab="NOx(1000toneladas)",xlab="MIO-KVM",main="NOx Emission vs Distance (France)")
lines(sort(x),cbind(1,sort(x.scaled),sort(x.scaled)^2,sort(x.scaled)^3)%*%coef.pol3*y.sd+y.mean)

# Predicción(punto rojo en la figura)
newPoint <- 618266 # nuevo x 
newPoint.scaled <- (newPoint-x.mean)/x.sd
points(newPoint,cbind(1,newPoint.scaled,newPoint.scaled^2,newPoint.scaled^3)%*%coef.pol3*y.sd+y.mean,col="red",pch=16)
```
