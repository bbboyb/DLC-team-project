```r
# Modelo de regresión (grado2 y grado3)
coefficients_matrix_pol2 <- function(x0,y){
    x <- cbind(1,x0,x0^2)
    return(solve(t(x)%*%x)%*%t(x)%*%y)
}
coefficients_matrix_pol3 <- function(x0,y){
    x <- cbind(1,x0,x0^2,x0^3)
    return(solve(t(x)%*%x)%*%t(x)%*%y)
}
```

```r
# Regresión: Francia
y <- data[1:9,3] # Y es la variable objetivo: Emisión de NOx
x <- data[1:9,5] # X es la variable independiente: Distancia recorrida
x.mean <- mean(x)
x.sd <- sd(x)
y.mean <- mean(y)
y.sd <- sd(y)

x.scaled <- scale(x)
y.scaled <- scale(y)
# Coeficientes de la regresión:
coef.pol3 <- coefficients_matrix_pol3(x.scaled,y.scaled)
plot(x,y,pch=16,ylab="NOx(1000toneladas)",xlab="MIO-KVM",main="NOx Emission vs Distance (France)")
lines(sort(x),cbind(1,sort(x.scaled),sort(x.scaled)^2,sort(x.scaled)^3)%*%coef.pol3*y.sd+y.mean)
newPoint <- 618266 # nuevo x 
newPoint.scaled <- (newPoint-x.mean)/x.sd
points(newPoint,cbind(1,newPoint.scaled,newPoint.scaled^2,newPoint.scaled^3)%*%coef.pol3*y.sd+y.mean,col="red",pch=16)
```

```r
# Cargar librerías necesarias
library(randomForest)
library(dplyr)
```

```r
#cambiamos los nombres de las columnas
colnames(data) <- c("X","YEAR","NOx_emission","NOx_reduction","MIO.VKM","EV","Country")
head(data)
```

```r
#separo el dataset por pais
data_france <- data %>% filter(Country == "France")
data_netherlands <- data %>% filter(Country == "Netherlands")
data_sweden <- data %>% filter(Country == "Sweden")
data_spain <- data %>% filter(Country == "Spain")
```

```r
dim(data_spain)
dim(data_sweden)
dim(data_france)
dim(data_netherlands)

head(data_netherlands)
```

```r



# Dividir datos en conjunto de entrenamiento y prueba (80%-20%)
set.seed(42)
indtrain <- sample(seq_len(nrow(data_spain)), size = 0.8 * nrow(data_spain))
train_spain <- data_spain[indtrain, ]
test_spain <- data_spain[-indtrain, ]

train_france <- data_france[indtrain, ]
test_france <- data_france[-indtrain, ]

train_netherlands <- data_netherlands[indtrain, ]
test_netherlands <- data_netherlands[-indtrain, ]

train_sweden <- data_sweden[indtrain, ]
test_sweden <- data_sweden[-indtrain, ]


# Entrenar modelo de regresión con Random Forest
model_rf_spain <- randomForest(NOx_emission ~ MIO.VKM, data = data_spain, ntree = 100)

model_rf_france <- randomForest(NOx_emission ~ MIO.VKM, data = data_france, ntree = 100)

model_rf_netherlands <- randomForest(NOx_emission ~ MIO.VKM, data = data_netherlands, ntree = 100)

model_rf_sweden <- randomForest(NOx_emission ~ MIO.VKM, data = data_sweden, ntree = 100)

#predecir
pred_spain <- predict(model_rf_spain, test_spain)
pred_france <- predict(model_rf_france, test_france)
pred_netherlands <- predict(model_rf_netherlands, test_netherlands)
pred_sweden <- predict(model_rf_sweden, test_sweden)

# Evaluar el modelo

acc.class <- function(pred, actual) {
    return(sum(diag(table(pred, actual))) / length(actual))
}


accuracy_spain = acc.class(pred_spain, test_spain$NOx_emission)
mse_spain <- mean((pred_spain - test_spain$NOx_emission)^2)
r2_spain <- 1 - (sum((test_spain$NOx_emission - pred_spain)^2) / sum((test_spain$NOx_emission - mean(test_spain$NOx_emission))^2))

accuracy_france = acc.class(pred_france, test_france$NOx_emission)
mse_france <- mean((pred_france - test_france$NOx_emission)^2)
r2_france <- 1 - (sum((test_france$NOx_emission - pred_france)^2) / sum((test_france$NOx_emission - mean(test_france$NOx_emission))^2))

accuracy_netherlands = acc.class(pred_netherlands, test_netherlands$NOx_emission)
mse_netherlands <- mean((pred_netherlands - test_netherlands$NOx_emission)^2)
r2_netherlands <- 1 - (sum((test_netherlands$NOx_emission - pred_netherlands)^2) / sum((test_netherlands$NOx_emission - mean(test_netherlands$NOx_emission))^2))

accuracy_sweden = acc.class(pred_sweden, test_sweden$NOx_emission)
mse_sweden <- mean((pred_sweden - test_sweden$NOx_emission)^2)
r2_sweden <- 1 - (sum((test_sweden$NOx_emission - pred_sweden)^2) / sum((test_sweden$NOx_emission - mean(test_sweden$NOx_emission))^2))


```

```r
print(paste("Mis predicciones se desvian el MSE Spain ", sqrt(mse_spain)))
print(paste("R2 Spain: El modelo explica apox el ", r2_spain*100,"% de la variabilidad de las emisiones de NOx en España"))
print(paste("El accuracy del modelo es ", accuracy_spain*100,"%"))
print("")
print(paste("Mis predicciones se desvian el MSE France ", sqrt(mse_france)))
print(paste("R2 France: El modelo explica apox el  ", r2_france*100,"% de la variabilidad de las emisiones de NOx en Francia"))
print(paste("El accuracy del modelo es ", accuracy_france*100,"%"))
print("")
print(paste("Mis predicciones se desvian el MSE Netherlands ", sqrt(mse_netherlands)))
print(paste("R2 Netherlands: El modelo explica apox el  ", r2_netherlands*100,"% de la variabilidad de las emisiones de NOx en Netherlands"))
print(paste("El accuracy del modelo es ", accuracy_netherlands*100,"%"))
print("")
print(paste("Mis predicciones se desvian el MSE Sweden ", sqrt(mse_sweden)))
print(paste("R2 Sweden : El modelo explica apox el ", r2_sweden*100,"% de la variabilidad de las emisiones de NOx en Sweden"))
print(paste("El accuracy del modelo es ", accuracy_sweden*100,"%"))
```

```r
# Crear un DataFrame con las métricas de evaluación
metrics <- data.frame(
  Country = c("Spain", "France", "Netherlands", "Sweden"),
  MSE = c(mse_spain, mse_france, mse_netherlands, mse_sweden),
  R2 = c(r2_spain, r2_france, r2_netherlands, r2_sweden)
)

# Gráfico de MSE
barplot(metrics$MSE, 
        names.arg = metrics$Country,
        col = "skyblue", 
        main = "Comparación del MSE entre los países",
        xlab = "País",
        ylab = "MSE (Error Cuadrático Medio)",
        ylim = c(0, max(metrics$MSE) * 1.1))

# Gráfico de R^2
barplot(metrics$R2, 
        names.arg = metrics$Country,
        col = "lightgreen", 
        main = "Comparación del R^2 entre los países",
        xlab = "País",
        ylab = "R^2 (Coeficiente de Determinación)",
        ylim = c(0, 1))  # El R² siempre estará entre 0 y 1
```
