```r
# Dataset1: NOx-PM2.5.csv, contiene los datos de NOx y PM2.5 para los 5 paises
data <- read.csv("NOx-PM2.5.csv",header=TRUE)
head(data)

# He sacado los datos correspondientes a los sectores de transporte
transport.sector.road <- data[which(data$EEA.Sub.Sector=="Road-transport"),]
transport.sector.nroad <- data[which(data$EEA.Sub.Sector=="Non-road transport"),]


NOx.road <- transport.sector.road[1:5,]
NOx.nroad <- transport.sector.nroad[1:5,]
PM2.5.road <- transport.sector.road[6:10,]
PM2.5.nroad <- transport.sector.nroad[6:10,]

```

```r
data3 <- read.csv("battery-plugin-hybrid-vehicles.csv",header=TRUE)
Countries <- c("France","Netherlands","Romania","Spain","Sweden")
vehicles <- data3[(data3$Entity %in% Countries) & (data3$Year >= 2014 & data3$Year <= 2022), ]
vehicles$Total_share <- vehicles$battery_electric_share + vehicles$plugin_hybrid_share
plot(vehicles$Year,vehicles$Total_share,col=as.factor(vehicles$Entity),pch = 19)
legend("topright", legend = unique(vehicles$Entity), col = unique(as.factor(vehicles$Entity)), pch = 19)
```
